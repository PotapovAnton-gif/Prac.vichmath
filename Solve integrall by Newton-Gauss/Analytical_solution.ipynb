$\large{Вычислить \space интеграл}$

$$I = \int_0^1{\frac{sin(100x)}{1 + x} dx}$$

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.special as sc
```

```python
def f(x):
    return np.sin(100*x)/(1 + x)

epsilon = 1e-6
```

```python
x = np.arange(0, 1, 0.0001)
y = f(x)

ax = plt.subplot()
ax.plot(x, y)
plt.grid()
plt.show
```

Вычислим h из формулы:

$$\epsilon \leqslant \frac{1}{2} \max\limits_{a \leqslant x \leqslant b}|f'(x)|(b - a)h$$


```python
h = 2e-6
x = np.arange(0, 1, h)
I = 0
for i in x:
    I += f(i)*h
print(I)
```
