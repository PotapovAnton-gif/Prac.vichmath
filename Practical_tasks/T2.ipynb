$\Large{Вычислить \; интеграл}$

$I = \int\limits_a^b sin(100x) \cdot exp(-x^2)\cdot cos(2x) dx$

$f(x) = sin(100x) \cdot exp(-x^2)\cdot cos(2x)$

```python
import math
import numpy as np 
import scipy.special as sp
import matplotlib.pyplot as plt
```

```python
def f(x):
    return np.sin(100*x)*np.exp(-(x**2)) * np.cos(2*x) 
```

```python
x_es = np.arange(0, 3, 0.00001)
y_es = f(x_es)

fig, ax = plt.subplots()
ax.plot(x_es, y_es)
ax.grid()

plt.show()
```

Из графика видно, как сильно асцилирует функция на практически на всей области значений $\to $ воспользумеся квадратурными формулами Гауса

$\int\limits_{-1}^{1}f(x) dx = \sum\limits_{i = 1}^{n}c_if(x_i) $

Произведем замену переменной: 

$\left\{ \begin{gathered} a*0 + b = -1 \\ a*3 + b = 1 \end{gathered}\right.  \to \left\{ \begin{gathered} b = -1 \\ a = 2/3 \end{gathered}\right. \to \phi = \frac{2}{3}*x - 1$

тогда $g(\phi)$ : $\left\{ \begin{gathered} x = \frac{3}{2} (phi + 1) \\ g(phi) = sin(100\frac{3}{2} (phi + 1)) \cdot exp(-\frac{3}{2} (phi + 1)^2)\cdot cos(2\frac{3}{2} (phi + 1)) \end{gathered}\right.$

```python
def g(phi):
    return np.sin(100*(3/2 * (phi + 1)))*np.exp(-((3/2 * (phi + 1))**2)) * np.cos(2*(3/2 * (phi + 1)))
```

```python
phi_es = 2/3 * x_es - 1
new_y_es = g(phi_es)

fig, ac = plt.subplots()
ac.plot(phi_es, new_y_es)
ac.grid()

plt.show()
```

После замены переменной, получим: 

$\int\limits_0^3 f(x) dx = \int\limits_{-1}^1 g(\phi)d\phi$

```python
dots, weght = sp.roots_legendre(100)
Integral = 0
for i in range(100):
    Integral += weght[i]*g(dots[i])
print(Integral)
```
